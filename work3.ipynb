```python
import numpy as np
import sys
import cv2
from google.colab.patches import cv2_imshow

# 画像の取得
gazou = input('ドット絵にしたい画像を拡張子も含めて入力してください')
img = cv2.imread(gazou)

if img is None:
  sys.exit('エラー：ファイルを開けませんでした')

if (img.shape[0] % 10) != 0 or (img.shape[1] % 10) != 0:
  img_h = img.shape[0] // 10 * 10
  img_w = img.shape[1] // 10 * 10
  img = cv2.resize(img,(img_w,img_h))
cv2_imshow(img)



h = img.shape[0]
w = img.shape[1]

hh = h / 10
ww = w / 10

print(h)
print(w)

for i in range(int(hh * ww)):
  yoko = int((i%ww) * 10)
  tate = int((i//ww) * 10)
  avg_b = 0
  avg_g = 0
  avg_r = 0
  for iy in range(tate,tate+10):
    for ix in range(yoko,yoko+10):
      try:
        avg_b += img[iy,ix][0]
        avg_g += img[iy,ix][1]
        avg_r += img[iy,ix][2]
      except IndexError:
        pass
  avg_b /= 100
  avg_g /= 100
  avg_r /= 100
  cv2.rectangle(img,(yoko,tate),(yoko+10,tate+10),color=(avg_b,avg_g,avg_r),thickness=-1)

cv2_imshow(img)


```
