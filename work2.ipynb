```python
# マインスイーパー（正方形）
import numpy as np
mass = int(input('縦横のマス数を2～18から指定してください：'))
board = np.zeros((mass,mass))
board2 = np.where(board == 0,'□',0) # 盤面の作成
board3 = board2.copy() # 爆弾の可視化

print(board2)

bomb_list = []
bomb = int(input('爆弾の数は？:'))
for i in range(bomb):
  a = np.random.randint(0,mass)
  b = np.random.randint(0,mass)
  if [a,b] not in bomb_list:
    board3[a][b] = '💣'
    bomb_list.append([a,b])
  else:
    if len(bomb_list) != i:
      board3[np.random.randint(0,mass)][np.random.randint(0,mass)] = '💣'
      bomb_list.append([np.random.randint(0,mass),np.random.randint(0,mass)])
    continue

#print(board2)
#print(board3)

print('注：フラグを爆弾がない場所に立てるとゲームオーバーになります。')
print('注：また、このプログラムは不完全のため、一番上の列を指定して調べると一番下の列に一番上の列のヒントが、一番左の列を指定して調べると一番右の列に一番左の列のヒントが表示されるようになっています。(例：ドラゴンクエストのワールドマップ)')
while len(bomb_list) > 0:
  mode = int(input('モードを選択してください\n １．爆弾を探す\n ２．フラグを立てる'))
  if mode == 1:
    x = int(input('調べたい場所の横の座標を指定してください。\n 左を0とします。'))
    y = int(input('調べたい場所の縦の座標を指定してください。\n 上を0とします。'))
    if board3[y][x] == '💣':
      print('爆弾が爆発しました。ゲームオーバーです。')
      break
    else:
      board2[y][x] = ' '
      Topleft = 0
      Topcenter = 0
      Topright = 0
      Left = 0
      Right = 0
      Underleft = 0
      Undercenter = 0
      Underright = 0
      if board3[y-2][x-2] == '💣':
        Topleft += 1
      if board3[y-2][x-1] == '💣':
        Topleft += 1
        Topcenter += 1
      if board3[y-2][x] == '💣':
        Topleft += 1
        Topcenter += 1
        Topright += 1
      if board3[y-2][x+1] == '💣':
        Topcenter += 1
        Topright += 1
      if board3[y-2][x+2] == '💣':
        Topright += 1
      if board3[y-1][x-2] == '💣':
        Topleft += 1
        Left += 1
      if board3[y-1][x-1] == '💣':
        Topcenter += 1
        Left += 1
      if board3[y-1][x] == '💣':
        Topleft += 1
        Left += 1
        Topright += 1
        Right += 1
      if board3[y-1][x+1] == '💣':
        Topcenter += 1
        Right += 1
      if board3[y-1][x+2] == '💣':
        Topright += 1
        Right += 1
      if board3[y][x-2] == '💣':
        Topleft += 1
        Left += 1
        Underleft += 1
      if board3[y][x-1] == '💣':
        Topleft += 1
        Topcenter += 1
        Underleft += 1
        Undercenter += 1
      if board3[y][x+1] == '💣':
        Topcenter += 1
        Topright += 1
        Undercenter += 1
        Underright += 1
      if board3[y][x+2] == '💣':
        Topright += 1
        Right *= 1
        Underright += 1
      if board3[y+1][x-2] == '💣':
        Left += 1
        Underleft += 1
      if board3[y+1][x-1] == '💣':
        Left += 1
        Undercenter += 1
      if board3[y+1][x] == '💣':
        Left += 1
        Underleft += 1
        Right += 1
        Underright += 1
      if board3[y+1][x+1] == '💣':
        Right += 1
        Undercenter += 1
      if board3[y+1][x+2] == '💣':
        Right += 1
        Underright += 1
      if board3[y+2][x-2] == '💣':
        Underleft += 1
      if board3[y+2][x-1] == '💣':
        Underleft += 1
        Undercenter += 1
      if board3[y+2][x] == '💣':
        Underleft += 1
        Undercenter += 1
        Underright += 1
      if board3[y+2][x+1] == '💣':
        Undercenter += 1
        Underright += 1
      if board3[y+2][x+2] == '💣':
        Underright += 1
      if board2[y-1][x-1] != ' ':
        board2[y-1][x-1] = str(Topleft)
      if board2[y-1][x] != ' ':
        board2[y-1][x] = str(Topcenter)
      if board2[y-1][x+1] != ' ':
        board2[y-1][x+1] = str(Topright)
      if board2[y][x-1] != ' ':
        board2[y][x-1] = str(Left)
      if board2[y][x+1] != ' ':
        board2[y][x+1] = str(Right)
      if board2[y+1][x-1] != ' ':
        board2[y+1][x-1] = str(Underleft)
      if board2[y+1][x] != ' ':
        board2[y+1][x] = str(Undercenter)
      if board2[y+1][x+1] != ' ':
        board2[y+1][x+1] = str(Underright)
      print(board2)
  elif mode == 2:
    x = int(input('フラグを立てる横の座標を指定してください。\n 左を0とします。'))
    y = int(input('フラグを立てる縦の座標を指定してください。\n 上を0とします。'))
    board2[y][x] = '🚩'
    if board3[y][x] != '💣':
      print('間違った場所にフラグを立てました。\n ゲームオーバーです。')
      break
    bomb_list.pop()
    print(board2)
    if len(bomb_list) == 0:
      print('すべての爆弾の位置が分かりました。ゲームクリアです。')
      break
  else:
    continue








```
