```python
# Cord1
import pandas as pd
import random
from matplotlib import pyplot as plt
import seaborn as sns

a = [] #データを手入力する場合とランダムでデータ生成する場合にデータを入れるからのリストaの作成
choice = int(input('データを手入力する場合は１、ランダムでデータを生成する場合は２、データの入ったcsvファイルを読み込む場合は３を入力してください')) #どの方法でデータを入れるかの確認
if choice == 1:
  count = int(input('何人分のデータを入力しますか？'))
  for i in range(count):
    graphic = int(input('映像の評価を１～４で入力してください'))
    sound = int(input('音楽の評価を１～４で入力してください'))
    character = int(input('登場人物の評価を１～４で入力してください'))
    story = int(input('物語の評価を１～４で入力してください'))
    battle = int(input('戦闘の評価を１～４で入力してください'))
    score = int(input('統合的な評価を1～20で入力してください'))
    a.append([graphic,sound,character,story,battle,score])
    df = pd.DataFrame(a) #上のプログラムで生成したデータをデータフレームにする
elif choice == 2:
  count = int(input('何人分のデータを生成しますか？'))
  for i in range(count):
    graphic = random.randint(1,4)
    sound = random.randint(1,4)
    character = random.randint(1,4)
    story = random.randint(1,4)
    battle = random.randint(1,4)
    score = graphic+sound+character+story+battle
    a.append([graphic,sound,character,story,battle,score])
    df = pd.DataFrame(a) #上のプログラムで生成したデータをデータフレームにする
elif choice == 3:
  a = input('読み込むファイル名を入力してください')
  df = pd.read_csv(a)  #上のプログラムで読み取ったデータをデータフレームにする


df.columns = ['graphic','sound','character','story','battle','score']
df
```

```python
# Cord2
plt.hist(df['score'],range=(0,21),bins=21)#ヒストグラムの作成
plt.show()

sns.heatmap(df.corr()) #ヒートマップの作成
plt.show()

df.corr() #相関係数一覧の表示
```
